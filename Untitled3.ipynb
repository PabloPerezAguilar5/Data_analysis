```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('netflix1.csv')
```

```python
nulos_por_columna = df.isnull().sum()
total_nulos = df.isnull().sum().sum()
print("Valores nulos por columna:\n", nulos_por_columna)
print("Total de valores nulos en el dataset:", total_nulos)
```

```python
df.head()
```

```python
incompletos_por_columna = df.isna().sum()
total_incompletos = df.isna().sum().sum()
print("Valores incompletos por columna:\n", incompletos_por_columna)
print("Total de valores incompletos en el dataset:", total_incompletos)
```

```python
df['release_year'] = df['release_year'].astype(int)
print(df.dtypes['release_year']) 
```

```python
df['first_listed'] = df['listed_in'].apply(lambda x: x.split(',')[0])
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['release_year'], bins=20, color='blue', edgecolor='black')
plt.title('Distribución de Release Year')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Frecuencia')
plt.show()
```

```python
first_listed_counts = df['first_listed'].value_counts()

plt.figure(figsize=(12, 8))
first_listed_counts.plot(kind='bar', color='green')
plt.title('Frecuencia de Primeros Valores en Listed In')
plt.xlabel('Categoría')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
releases_per_year = df['release_year'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
plt.plot(releases_per_year.index, releases_per_year.values, marker='o', linestyle='-')
plt.title('Número de Lanzamientos por Año')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Número de Lanzamientos')
plt.grid(True)
plt.show()
```
